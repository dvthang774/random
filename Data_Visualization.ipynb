```python
x = df.groupby('Highest_Qualified_Member')['Highest_Qualified_Member'].count()
x = pd.DataFrame(x)
x.columns = ['Percent']
highest_qm = x/sum(x['Percent'])*100
highest_qm
```

```python
df_pie=highest_qm
_df_pie=highest_qm['Percent'].values.tolist()
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = [0,0.1,0,0,0.1]

plt.pie(_df_pie, colors = colors, autopct = '%1.1f%%', startangle = 90, explode = explode, pctdistance = 1.12, shadow = True)
plt.title('Percentage', color = 'black', y = 1.1, fontsize = 18)
plt.legend(df_pie.index, loc = 1, fontsize = 8)
plt.axis('equal')
# plt.savefig('pie.png')

plt.show()
```

```python
highest_qm.reset_index(inplace=True)
highest_qm
sns.barplot(y='Highest_Qualified_Member',x='Percent',data = highest_qm )
plt.title('Percentage', color = 'black', y = 1.1, fontsize = 18)
# plt.savefig('pie.png')
plt.show()
```
