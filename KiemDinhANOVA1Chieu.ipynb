```python
d1=data["X1"]
d2= data["X2"]
d3= data["X3"]
d4= data["X4"]
d5= data["X5"]
```

```python
proj_1=ta.proj2(d1)
proj_1.hist()
proj_1.qq()
proj_1.shapiro()
proj_1.kstest()
```

```python
proj_2=ta.proj2(d2)
# proj_1.standardize()
proj_2.hist()
proj_2.qq()
proj_2.shapiro()
proj_2.kstest()
```

```python
proj_5=ta.proj2(d5)
# proj_1.standardize()
proj_5.hist()
proj_5.qq()
proj_5.shapiro()
proj_5.kstest()
```

```python
#Code kiểm định tính đồng nhất về phương sai của các nhóm
stat, p_value = stats.bartlett(d1,d2,d3,d4,d5)
print('statistic by Bartlett = ',stat)
print('pvalue by Bartlett = ',p_value)
if (p_value > 0.05):
    print('Không đủ bằng chứng để bác bỏ: Đây là các dữ liệu đồng nhất')
else:
    print('Đấy không phải là các dữ liệu đồng nhất')
```

```python
#kiểm định anova 1 chiều
from scipy.stats import f_oneway
statistic,pvalue = f_oneway(d1,d2,d3,d4,d5)
print("Kiểm định anova 1 chiều: \n statistic = ",statistic,"\n pvalue = ",pvalue)
```
